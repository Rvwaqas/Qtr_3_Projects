```python
import random

def main():
    choice_list = ["rock", "paper", "scissors"]
    guess = random.choice(choice_list)
    user_guess = input("Choose rock, paper, or scissors: ").lower()
    if user_guess not in choice_list:
        print("Invalid Value enter")
        return  # Exit the function if the input is invalid

    # Compare user_guess with guess (computer's choice)
    if user_guess == guess:
        print("It's a tie!")
    elif (user_guess == "rock" and guess == "scissors") or \
         (user_guess == "paper" and guess == "rock") or \
         (user_guess == "scissors" and guess == "paper"):
        print("You win!")
    else:
        print("You lose!")

def play():
    while True:
        main()
        play_again = input("Play again? Type 'yes' or 'no': ")
        # Changed to != to match the description
        if play_again != 'yes':
            break

play()
```
