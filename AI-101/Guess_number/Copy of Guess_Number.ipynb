```python
import random


guess_number=random.randint(1,100)
chance=0

while (chance<3):
  user_number=int(input("Enter you Guess number In between 1 to 100: "))
  if(user_number==guess_number):
    print(f"You win and guess number is {guess_number} and your number is {user_number}")
    break
  else:
    print(f"you lost guess_number is {guess_number}")
    chance+=1



```
